```python
import numpy as np
import matplotlib.pyplot as plt

n = int(input('Supply an integer for analysis:' )) #np.random.randint(10)
count, x, y = 0, [0], [n]

def collatz(z):
	# Unification of Collatz Conjecture rule: x is odd  : 3x+1 or x is even : x/2
	return 0.25*(2 + 7*z + (-1)*(2 + 5*z)*((-1) ** z))

while y[-1] != 1.0:
	count += 1; x.append(count); y.append(collatz(y[-1])) #; print(y[-1], end=", ")

print(f"Intial values: {n} \nNumber of Iterations: {len(y)-1} \nSequence values: {y}")

def plt_def(data, xlabel='',ylabel='',plttitle='',pltsave=''):
    plt.xlabel(xlabel); plt.ylabel(ylabel); plt.title(plttitle); plt.savefig(pltsave, dpi=1000, format='png'); plt.show()

plt_def(plt.plot(x,y), f'Number of Iterations ({len(y)-1})', f'Sequence Values ({n})', f'Graphical representation of Collatz Conjecture for n={n}', f'../results/images/Collatz_Conjecture(n={n}).png')
```
